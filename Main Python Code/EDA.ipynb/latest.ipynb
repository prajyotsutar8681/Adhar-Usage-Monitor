```python
enrol_df = pd.concat([pd.read_csv(f) for f in enrolment_files], ignore_index=True)
demo_df  = pd.concat([pd.read_csv(f) for f in demo_files], ignore_index=True)
bio_df   = pd.concat([pd.read_csv(f) for f in bio_files], ignore_index=True)
```

```python
print("DEMOGRAPHIC DATA SHAPE:", demo_df.shape)
display(demo_df.head())
```

```python
print("BIOMETRIC DATA SHAPE:", bio_df.shape)
display(bio_df.head())
```

```python
def clean_region_names(df):
    for col in ["state", "district"]:
        if col in df.columns:
            df[col] = df[col].astype(str).str.strip().str.title()
    return df
```

```python
def add_year_column(df):
    df["date"] = pd.to_datetime(df["date"], dayfirst=True, errors="coerce")
    df["year"] = df["date"].dt.year
    return df
```

```python
enrol_df = clean_region_names(enrol_df)
demo_df = clean_region_names(demo_df)
bio_df = clean_region_names(bio_df)

enrol_df = add_year_column(enrol_df)
demo_df = add_year_column(demo_df)
bio_df = add_year_column(bio_df)

enrol_df.drop_duplicates(inplace=True)
demo_df.drop_duplicates(inplace=True)
bio_df.drop_duplicates(inplace=True)
```

```python
print(enrol_df.columns.tolist())
```

```python
enrol_yearly = enrol_df.groupby(
    "year", as_index=False
)[["age_5_17", "age_18_greater"]].sum()

enrol_long = enrol_yearly.melt(
    id_vars="year",
    var_name="age_group",
    value_name="enrolment_count"
)

enrol_long["age_group"] = enrol_long["age_group"].map({
    "age_5_17": "Age 5–17",
    "age_18_greater": "Age 18+"
})

```

```python
print(enrol_long.head())
print(enrol_long.isna().sum())
```

```python
fig_enrol = px.line(
    enrol_long,
    x="year",
    y="enrolment_count",
    color="age_group",
    markers=True,
    title="Aadhaar Enrolment Growth by Age Group"
)
fig_enrol.show()
```

```python
print(bio_df.columns.tolist())

```

```python
bio_yearly = bio_df.groupby(
    "year", as_index=False
)[["bio_age_5_17", "bio_age_17_"]].sum()

bio_long = bio_yearly.melt(
    id_vars="year",
    var_name="age_group",
    value_name="biometric_usage"
)

bio_long["age_group"] = bio_long["age_group"].map({
    "bio_age_5_17": "Age 5–17",
    "bio_age_17_": "Age 17+"
})
```

```python
fig_bio = px.line(
    bio_long,
    x="year",
    y="biometric_usage",
    color="age_group",
    markers=True,
    title="Aadhaar Biometric Usage Growth by Age Group"
)
fig_bio.show()
```

```python
common_age_groups = ["Age 5–17", "Age 17+"]

enrol_aligned = enrol_long[enrol_long["age_group"].isin(common_age_groups)]
bio_aligned = bio_long[bio_long["age_group"].isin(common_age_groups)]
```

```python
enrol_compare = enrol_long.rename(columns={"enrolment_count": "value"})
enrol_compare["type"] = "Enrolment"

bio_compare = bio_long.rename(columns={"biometric_usage": "value"})
bio_compare["type"] = "Biometric Usage"

gap_df = pd.concat([enrol_compare, bio_compare], ignore_index=True)
```

```python
fig_gap = px.line(
    gap_df,
    x="year",
    y="value",
    color="type",
    facet_col="age_group",
    markers=True,
    title="Enrolment vs Biometric Usage (Log Scale)"
)

fig_gap.update_yaxes(type="log")
fig_gap.show()


```

```python
state_year_age = enrol_df.groupby(
    ["state", "year"], as_index=False
)[["age_5_17", "age_18_greater"]].sum()

state_long = state_year_age.melt(
    id_vars=["state", "year"],
    var_name="age_group",
    value_name="enrolment_count"
)

state_long["age_group"] = state_long["age_group"].map({
    "age_5_17": "Age 5–17",
    "age_18_greater": "Age 18+"
})

```

```python
fig_state = px.bar(
    state_long,
    x="enrolment_count",
    y="state",
    color="age_group",
    animation_frame="year",
    orientation="h",
    title="State-wise Aadhaar Enrolment Race (Age-wise)",
    height=650
)

fig_state.update_layout(
    xaxis_title="Total Enrolments",
    yaxis_title="State",
    showlegend=True
)

fig_state.show()
```

```python
heat_df = state_long.groupby(
    ["state", "year"], as_index=False
)["enrolment_count"].sum()

fig_state = px.density_heatmap(
    heat_df,
    x="year",
    y="state",
    z="enrolment_count",
    color_continuous_scale="Viridis",
    title="State-wise Aadhaar Enrolment Intensity"
)

fig_state.show()
```

```python
enrol_long = enrol_long.sort_values(["age_group", "year"])

enrol_long["yoy_growth_pct"] = enrol_long.groupby(
    "age_group"
)["enrolment_count"].pct_change() * 100

anomalies = enrol_long[
    enrol_long["yoy_growth_pct"].abs() > 30
]

print("ANOMALIES (YoY Change > 30%)")
display(anomalies)
```

```python
usage_index = gap_df.pivot_table(
    index=["year", "age_group"],
    columns="type",
    values="value",
    aggfunc="sum"
).reset_index()

usage_index["usage_per_enrolment"] = (
    usage_index["Biometric Usage"] /
    usage_index["Enrolment"]
)

fig = px.line(
    usage_index,
    x="year",
    y="usage_per_enrolment",
    color="age_group",
    markers=True,
    title="Aadhaar Usage Intensity Index (Usage per Enrolment)"
)

fig.show()
```

```python
cum_df = enrol_long.copy()
cum_df["cumulative"] = cum_df.groupby(
    "age_group"
)["enrolment_count"].cumsum()

fig = px.area(
    cum_df,
    x="year",
    y="cumulative",
    color="age_group",
    title="Cumulative Aadhaar Enrolment Penetration"
)

fig.show()

```
